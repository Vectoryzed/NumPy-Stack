```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# linspace -> Generate data with start, end, number of points in between
x = np.linspace(0, 10, 10)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("Cool chart")
plt.plot(x, y)
```

```python
# Chart is choppy. This is due to number of points in the plot
# Repeat with more points
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("Cool chart")
plt.plot(x, y)
```
