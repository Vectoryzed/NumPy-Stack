- Fourier transform (FFT) - In **numpy**, converts signal from time domain to frequency domain

```python
x = np.linspace(0, 100, 10000)
y = np.sin(x) + np.sin(3*x) + np.sin(5*x)
plt.plot(y)
```

```python
# FFT 
# FFT gives us signal with complex numbers so we find magnitude before plotting
Y = np.fft.fft(y)
plt.plot(np.abs(Y))
```

```python
# Spikes at the front are at 16, 48 and 80; these are the original frequences
print(2*np.pi*16/100)
print(2*np.pi*48/100)
print(2*np.pi*80/100)
# These results are in the original y = equation above
```
