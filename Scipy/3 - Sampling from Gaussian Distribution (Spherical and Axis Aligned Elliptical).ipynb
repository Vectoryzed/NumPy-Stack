```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
r = np.random.randn(10000, 2)
```

```python
# Normal distribution on multiple axis are circular (2D) or spherical (3D)
plt.scatter(r[:,0], r[:,1])
```

```python
# Want an ellipitcal gaussian
r[:,1] = 5*r[:,1]+2
plt.scatter(r[:,0], r[:,1])
# Notice the Scale of Y compared to above
```

```python
# Remove autoscaling
plt.axis('equal')
plt.scatter(r[:,0], r[:,1])
```
